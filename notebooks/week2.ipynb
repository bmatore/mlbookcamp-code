```python
import numpy as np
import pandas as pd
 
from matplotlib import pyplot as plt
import seaborn as sns
%matplotlib inline

sns.set(rc={'figue.figsize':(11.7,8.27)})
```

```python
# read data
df = pd.read_csv('/home/bmatore/projects/mlbookcamp-code/chapter-02-car-price/data.csv')
df.head()
```

```python
# normalise column names and features
df.columns = df.columns.str.lower().str.replace(' ', '_')
 
string_columns = list(df.dtypes[df.dtypes == 'object'].index)
 
for col in string_columns:
    df[col] = df[col].str.lower().str.replace(' ', '_')
```
